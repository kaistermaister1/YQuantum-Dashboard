```python
import pulp
import numpy as np

def make_C_matrix(problem):

    M=problem.M
    N=problem.N

    x = {}
    for i in range(N):
            for m in range(M):
                x[i, m] = pulp.LpVariable(f"x_{i}_{m}", cat=pulp.LpBinary)

    coverages = problem.coverages
    beta = problem.price_sensitivity_beta
    objective_terms = np.zeros(N,M)

    for m in range(M):
        delta_m = problem.get_discount(m)
        for i in range(N):
            cov = coverages[i]
            alpha_im = problem.get_affinity(i, m)
            margin = (
                cov.price
                * cov.contribution_margin_pct
                * (1 - delta_m)
                * cov.take_rate
                * alpha_im
                * (1 + beta * delta_m)
            )
            objective_terms[i,m] = margin
    return objective_terms
```
